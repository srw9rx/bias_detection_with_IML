```python
import pandas as pd
import glob
import os
  
# merging the files
joined_files = os.path.join("/Users/sophiawalton/Documents/CS6501", "newsarticleswtext*.tsv")
  
# A list of all joined files is returned
joined_list = glob.glob(joined_files)
print(joined_list)
joined_list.remove('/Users/sophiawalton/Documents/CS6501/newsarticleswtextusatoday.tsv')
df = pd.read_csv('/Users/sophiawalton/Documents/CS6501/newsarticleswtextusatoday.tsv', sep='\t', header=0)
for path in joined_list:
    tsv = pd.read_csv(path, sep='\t')
    df = pd.concat((df, tsv))
 

df.to_csv('/Users/sophiawalton/Documents/CS6501/newsarticleswtext.tsv', sep='\t')
```

```python
mediabias = pd.read_csv('/Users/sophiawalton/Documents/CS6501/newsarticleswtext.tsv', sep='\t', header=0)
del mediabias['Unnamed: 0']
del mediabias['Unnamed: 0.1']
print(mediabias.head())
print(mediabias['democrat.vote'].unique())
newlabels = []
for article in mediabias.iterrows():
    #print(len(article[1]))
    val = abs(float(article[1][2]))
    if val == 1:
        newlabels.append('1')
    #elif article[1][5].find('Somewhat') != -1:
     #   newlabels.append('1')
    else:
        newlabels.append('0')
mediabias['new_labels'] = newlabels
mediabias = mediabias[mediabias.text != '400'] #drop page not found errors
mediabias2 = mediabias.dropna()
print(mediabias2.head())
print(mediabias2.tail())
```

```python
import pandas as pd

MBIC = pd.read_csv('/Users/sophiawalton/Documents/CS6501/labeled_dataset.tsv', sep='\t', header=0)
del MBIC['Unnamed: 0']
print(MBIC.head())
newlabels = []
for article in MBIC.iterrows():
    #print(article[1][7])
    #print(len(article[1]))
    #if article[1][8].lower().find("somewhat") != -1:
    #    newlabels.append('1')
    if article[1][7] == 'Non-biased':
        newlabels.append('0')
    else:
        newlabels.append('1')
MBIC['new_labels'] = newlabels
MBIC2 = MBIC.dropna()
print(MBIC2.head())
print(MBIC2.tail())
```
